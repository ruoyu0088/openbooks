## 将Cython程序分多个单元定义

```python
%load_ext cythonmagic
```

```python
%%cython_part func_add
%%cython
cdef int add(int a, int b):
    return a + b
```

```python
%%cython_part main
def add_list(list a, list b):
    cdef int i
    cdef list c = []
    for i in range(len(a)):
        c.append(add(a[i], b[i]))
    return c
```

```python
%run_cython_parts func_add main
```

```python
add_list([1,2,3],[3,4,5])
```
