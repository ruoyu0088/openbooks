```python
import cv2
from cv2 import cv
%matplotlib inline
import pylab as pl
import numpy as np
import pandas as pd
```

# 把照片转换成圆形马赛克效果

首先使用`Canny()`找出图像的边缘，非边缘部分可以用大圆填充。

```python
img = cv2.imread("lena.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
edges = cv2.Canny(img, 50, 200)

fig, axes = pl.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(img)
axes[1].imshow(~edges, cmap="gray")
for ax in axes:
    ax.axis("off")
fig.subplots_adjust(0, 0, 1, 1, 0, 0)
```

用`distanceTransform()`找到每个白色像素到最近的黑色像素的距离。所有距离中的最大值就是能填充的最大的圆。找到这个最大可填充圆的圆形`(cx, cy)`和半径`radius`。

```python
edges2 = ~edges
dist = cv2.distanceTransform(edges2, cv.CV_DIST_L2, 3)

idx = np.argmax(dist)
cy, cx = np.unravel_index(idx, dist.shape)
radius = dist[cy, cx]

fig, axes = pl.subplots(1, 2, figsize=(8, 4))
axes[0].imshow(dist)
cv2.circle(edges2, (cx, cy), radius, (0, 0, 0, 0), thickness=-1)
axes[1].imshow(edges2, cmap="gray")

for ax in axes:
    ax.axis("off")
fig.subplots_adjust(0, 0, 1, 1, 0, 0)
```

`circle()`返回在图像大小为`shape`的图像上绘制圆心在`(cx, cy)`半径为`radius`的填充圆所填充的像素坐标。

```python
def circle(cx, cy, radius, shape):
    dr = 1.0 / float(radius)
    r, c = np.ogrid[-1:1:dr, -1:1:dr]
    rr, cc = np.where(r**2 + c**2 < 1)
    rr = rr + cy - int(radius)
    cc = cc + cx - int(radius)
    mask = (rr >= 0) & (rr < shape[0]) & (cc >= 0) & (cc < shape[1])
    return cc[mask], rr[mask]
```

```python
edges2 = ~edges
h, w = img.shape[:2]
circles = []

for i in range(500):
    dist = cv2.distanceTransform(edges2, cv.CV_DIST_L2, 3)
    idx = np.argmax(dist)
    cy, cx = np.unravel_index(idx, dist.shape)
    radius = dist[cy, cx]
    x, y = circle(cx, cy, int(radius), (h, w))
    edges2[y, x] = 0
    color = img[y, x].mean(0).astype(np.uint8)
    circles.append([cx, cy, radius] + color.tolist())
    
circles = np.array(circles)
```

为了快速绘制大量的圆形，这里使用`EllipseCollection`：

```python
from matplotlib.collections import EllipseCollection

def draw_circles(circles, scale=1.0, lw=0.1, ax=None):
    if ax is None:
        fig, ax = pl.subplots(figsize=(8, 8))
    ax.set_aspect("equal")
    
    circles = np.asarray(circles)
    diameters = circles[:, 2] * 2 * scale
    angles = np.zeros_like(diameters)
    centers = circles[:, :2]
    colors = circles[:, 3:] / 256.0

    ec = EllipseCollection(widths=diameters, heights=diameters, 
                           angles=angles, units="xy", 
                           offsets=centers, transOffset=ax.transData, 
                           facecolors=colors, edgecolors="w", lw=lw)
    ax.add_collection(ec)
    return ax
```

非边缘部分很好地被填充了，但是边缘部分没有圆形填充，这时可以去除边缘之后继续：

```python
fig, axes = pl.subplots(1, 3, figsize=(12, 4))
ax = draw_circles(circles, ax=axes[0])
axes[0].axis((0, w, h, 0));
axes[1].imshow(edges2.copy(), cmap="gray")
edges2[edges == 255] = 255
axes[2].imshow(edges2, cmap="gray")
for ax in axes:
    ax.axis("off")
fig.subplots_adjust(0, 0, 1, 1, 0, 0)
```

非边缘部分采用800个圆形填充，然后去除边缘中的黑色像素，再用700个圆填充：

```python
edges2 = ~edges
h, w = img.shape[:2]
circles = []

for i in range(1500):
    if i == 800:
        edges2[edges == 255] = 255    
    dist = cv2.distanceTransform(edges2, cv.CV_DIST_L2, 3)
    idx = np.argmax(dist)
    cy, cx = np.unravel_index(idx, dist.shape)
    radius = dist[cy, cx]
    x, y = circle(cx, cy, int(radius), (h, w))
    edges2[y, x] = 0
    color = img[y, x].mean(0).astype(np.uint8)
    circles.append([cx, cy, radius] + color.tolist())
    
circles = np.array(circles)
```

```python
fig, axes = pl.subplots(1, 2, figsize=(12, 6))
for ax in axes:
    ax.axis("off")
    ax.axis((0, w, h, 0))
fig.subplots_adjust(0, 0, 1, 1, 0, 0)

draw_circles(circles, ax=axes[0])
draw_circles(circles, ax=axes[1], scale=1.1);
```

用`EllipseCollection`画圆速度稍微有些慢，而`CircleCollection`不支持数据空间的圆形半径。因此我们编写下面的`DataCircleCollection`，其半径为数据空间中的长度。

```python
from matplotlib.collections import CircleCollection, Collection
from matplotlib.transforms import Affine2D

class DataCircleCollection(CircleCollection):
    
    def draw(self, render):
        ax = self.axes
        
        m = ax.transData.get_affine().get_matrix()
        ms = np.zeros((len(self._sizes), 3, 3))
        ms[:, 0, 0] = self._sizes
        ms[:, 1, 1] = self._sizes
        ms[:, 2, 2] = 1
        self._transforms = [Affine2D(m) for m in ms]
        
        m = ax.transData.get_affine().get_matrix().copy()
        m[:2, 2:] = 0
        self.set_transform(Affine2D(m))
        
        return Collection.draw(self, render)
    
def draw_circles2(circles, scale=1.0, lw=0.1, ax=None):
    if ax is None:
        fig, ax = pl.subplots(figsize=(8, 8))
    ax.set_aspect("equal")
    
    circles = np.asarray(circles)
    radius = circles[:, 2] * scale
    centers = circles[:, :2]
    colors = circles[:, 3:] / 256.0

    cc = DataCircleCollection(radius, 
                           offsets=centers, transOffset=ax.transData, 
                           facecolors=colors, edgecolors="white", lw=lw)
    ax.add_collection(cc)
    return ax
```

```python
ax = draw_circles2(circles, scale=1.1, lw=0.1)
ax.axis("off")
ax.axis((0, w, h, 0));
```
