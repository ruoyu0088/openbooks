```python
%matplotlib inline
import pylab as pl
import cv2
import numpy as np
```

# 自动计算接触角

```python
img = cv2.imread("contact_angle01.jpg")
h, w = img.shape[:2]
pl.imshow(img[:,:,::-1]);
```

界面以及水珠上面的背景模糊，如果进行边缘检测，不会出现边缘。对边缘图像寻找每列的第一个白色像素：

```python
th = 80
edge = cv2.Canny(img, th, th)
Y = edge.argmax(axis=0)
X = pl.arange(edge.shape[1])
pl.imshow(edge, cmap="gray")
pl.plot(X, Y, "r")
pl.axis([0, w, h, 0]);
```

用直线和圆构成的分段函数去拟合上面的红色点。

```python
def line_circle_intersect(b, cx, cy, r):
    dx = (r**2 - (b - cy)**2)**0.5
    x1 = cx - dx
    x2 = cx + dx
    return x1, x2
    
def line_circle(x, b, cx, cy, r):
    x1, x2 = line_circle_intersect(b, cx, cy, r)
    return np.where(x < x1, b, 
                    np.where(x > x2, b, 
                             cy - (r**2 - (cx - x)**2)**0.5))

Y2 = line_circle(X, 80, 100, 90, 70)

pl.plot(X, Y)
pl.plot(X, Y2)
ax = pl.gca()
ax.imshow(img)
ax.axis([0, w, h, 0]);
```

用`optimize.minimize()`找到最佳拟合参数：

```python
from scipy import optimize

def f(p):
    return np.abs(line_circle(X, *p) - Y).sum()

h, w = img.shape[:2]
res = optimize.minimize(f, (h/2, w/2, h/2, w/4))

Y2 = line_circle(X, *res.x)
pl.imshow(img)
pl.plot(X, Y2, "r", alpha=0.5);

b, cx, cy, r = res.x
angle = np.pi*0.5 - np.arcsin((cy - b) / r)

x1, x2 = line_circle_intersect(*res.x)

angle2 = np.pi - angle

length = 100
dy = -length*np.tan(angle2)
pl.plot([x2-length, x2+length], 
        [b-dy, b+dy], alpha=0.8)

ax = pl.gca()
ax.axis((0, w, h, 0));
```

```python
np.rad2deg(angle)
```

## 非水平直线

```python
img = cv2.imread("contact_angle02.jpg")
pl.imshow(img[:,:,::-1])
h, w = img.shape[:2]
```

```python
th = 80
edge = cv2.Canny(img, th, th)
Y = edge.argmax(axis=0)
X = pl.arange(edge.shape[1])
pl.imshow(edge, cmap="gray")
pl.plot(X, Y, "r")
pl.axis([0, w, h, 0]);
```

计算直线与圆的交点的函数变得比较复杂：

```python
def line_circle_intersect(k, b, cx, cy, r):
    from math import sqrt
    x0 = b - cy
    x1 = k**2 + 1
    x2 = cx - k*x0
    x3 = sqrt(-x1*(cx**2 - r**2 + x0**2) + x2**2)
    x4 = x2 + x3
    x5 = x2 - x3
    x6 = b*x1
    return [x5/x1, (k*x5 + x6)/x1, x4/x1, (k*x4 + x6)/x1]

def line_circle(x, k, b, cx, cy, r):
    try:
        x1, y1, x2, y2 = line_circle_intersect(k, b, cx, cy, r)
        x1, x2 = min(x1, x2), max(x1, x2)
        line = k * x + b
        circle = cy - (r**2 - (cx - x)**2)**0.5
        return np.where(x < x1, line, 
                        np.where(x > x2, line, circle))
    except ValueError:
        return np.zeros_like(x)
    
Y2 = line_circle(X, -0.1, 80.0, 100.0, 90.0, 60.0)

pl.plot(X, Y)
pl.plot(X, Y2)
ax = pl.gca()
ax.imshow(img)
ax.axis([0, w, h, 0]);
```

```python
def f(p):
    return np.abs(line_circle(X, *p) - Y).sum()

h, w = img.shape[:2]
res = optimize.minimize(f, (0, h/2, w/2, h/2, w/4))

Y2 = line_circle(X, *res.x)
pl.imshow(img)
pl.plot(X, Y2, "r", alpha=0.5);

k, b, cx, cy, r = res.x

x1, y1, x2, y2 = line_circle_intersect(*res.x)

if x2 < x1:
    x, y = x1, y1
else:
    x, y = x2, y2
    
angle = np.pi*0.5 - np.arcsin((cy - y) / r)
angle2 = np.pi - angle

length = 100
dy = -length*np.tan(angle2)
pl.plot([x-length, x+length], 
        [y-dy, y+dy], alpha=0.8)

ax = pl.gca()
ax.axis((0, w, h, 0));
```

```python
np.rad2deg(angle - np.arctan(k))
```

## SymPy公式推导

```python
from sympy.geometry import Line, Point, Circle
from sympy import symbols, cse
```

计算水平直线与圆的交点：

```python
b, k = symbols("b, k")
cx, cy = symbols("cx, cy")
r = symbols("r", positive=True)

line = Line(Point(0, b), slope=0)
circle = Circle(Point(cx, cy), r)
p1, p2 = line.intersection(circle)

print p1, p2
```

计算任意直线与圆的交点：

```python
line = Line(Point(0, b), slope=k)
circle = Circle(Point(cx, cy), r)
p1, p2 = line.intersection(circle)

print p1, p2
```

使用`cse()`可以将多个表达式转换成分步运算，然后将分步表达式输出成Python程序：

```python
steps, result = cse([p1.x, p1.y, p2.x, p2.y])
print "def line_circle_intersect(k, b, cx, cy, r):"
print "    from math import sqrt"
for step in steps:
    print "    %s = %s" % step
print "    return", result
```
